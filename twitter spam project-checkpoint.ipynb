## Create Word Cloud

###### With the help of countvectorizer we can Convert a collection of text documents to a matrix of token counts.
###### the number of features will be equal to the vocabulary size found by analyzing the data

```python
from sklearn.feature_extraction.text import CountVectorizer
ham_vect = CountVectorizer(stop_words='english')

spam_vect = CountVectorizer(stop_words='english')

ham_dtm = ham_vect.fit_transform(ham_data['SMS'])

spam_dtm = spam_vect.fit_transform(spam_data['SMS'])

```

```python
ham_word = ham_vect.get_feature_names()
ham_freqs = ham_dtm.sum(axis=0).A1
ham_result = dict(zip(ham_word, ham_freqs))
print(len(ham_word))
```

```python
spam_word = spam_vect.get_feature_names()
spam_freqs = spam_dtm.sum(axis=0).A1
spam_result = dict(zip(spam_word, spam_freqs))
print(len(spam_word))
```

```python
from wordcloud import WordCloud

#Initializing WordCloud using frequencies of tags.
wordcloud = WordCloud(    background_color='black',
                          width=1600,
                          height=800,
                    ).generate_from_frequencies(ham_result)

fig = plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
fig.savefig("ham_word_cloud.png")
plt.show()
```

### Above fig. is word cloud of a ham tweet

```python
wordcloud = WordCloud(    background_color='black',
                          width=1600,
                          height=800,
                    ).generate_from_frequencies(spam_result)

fig = plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
fig.savefig("spam_word_cloud.png")
plt.show()
```

### above fig is a wordcloud of spam tweet

## Split Data

```python
data['class'] = data['class'].replace('ham', 0)
data['class'] = data['class'].replace('spam', 1)
```

```python
data['class'].head()
```

###### we  can split data by using StratifiedShuffleSplit fun.

```python
x = data['SMS']
y = data['class']

from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits = 2, test_size = 0.20, random_state = 0)
sss.get_n_splits(x,y)
```

```python
for train_index, test_index in sss.split(x, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Tf-Idf

###### Convert a collection of raw documents to a matrix of TF-IDF features.

## for training set

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
train_dtm = tf.fit_transform(x_train)

print(tf.vocabulary_)
```

```python
type(train_dtm)
```

```python
train_data_array = train_dtm.toarray()
print(train_data_array)
print('Size of new array')

print(train_data_array.shape)
```

## for testing data

```python
test_dtm = tf.transform(x_test)
test_data_array = test_dtm.toarray()
print(test_data_array)
print('Size of new array')

print(test_data_array.shape)

#print(tf.vocabulary_)
```

## PCA

###### Linear dimensionality reduction using Singular Value Decomposition of the data to project it to a lower dimensional space.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(train_data_array)
pca_data = pca.transform(train_data_array)
main_data = pd.DataFrame(data = pca_data, columns=['component 1', 'component 2'])
#main_data.append(y_train)
main_data.head()
```

```python
sns.lmplot('component 1', 'component 2', data = main_data, fit_reg = False,)
plt.show()
```

# Classification Algorithm

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_data_array, y_train)
pr_lr = lr.predict(test_data_array)
```

```python
logistic_score = lr.score(test_data_array, y_test)
print(logistic_score)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, pr_lr))
print(classification_report(y_test, pr_lr))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors  = 3)
knn.fit(train_data_array, y_train)
pr_knn = knn.predict(test_data_array)
```

```python
knn_score = knn.score(test_data_array, y_test)
print(knn_score)
```

```python
print(confusion_matrix(y_test, pr_knn))
print(classification_report(y_test, pr_knn))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
gnb= GaussianNB()

gnb.fit(train_data_array, y_train, )
pr1=gnb.predict(test_data_array)
Naive_Bayes_Gaussian=gnb.score(train_data_array, y_train)
print(Naive_Bayes_Gaussian)
```

```python
print(confusion_matrix(y_test, pr1))
print(classification_report(y_test, pr1))
```

```python
mnb=MultinomialNB()
mnb.fit(train_data_array, y_train)
pr2=mnb.predict(test_data_array)
Naive_Bayes_MultiNomial=mnb.score(train_data_array, y_train)
print(Naive_Bayes_MultiNomial)
```

```python
print(confusion_matrix(y_test, pr2))
print(classification_report(y_test, pr2))
```

## SVM

```python
from sklearn.svm import SVC
sv=SVC(kernel='poly', degree=4)
sv.fit(train_data_array, y_train)
pr=sv.predict(test_data_array)
#print(pr)
SVM_poly_score=sv.score(test_data_array, y_test)
print(SVM_poly_score)
```

```python
print(confusion_matrix(y_test, pr))
print(classification_report(y_test, pr))
```

```python
sv2=SVC(kernel='rbf')
sv2.fit(train_data_array, y_train)
pr2=sv2.predict(test_data_array)
SVM_rbf_score=sv2.score(test_data_array, y_test)
print(SVM_rbf_score)
print(classification_report(y_test, pr2))
print(confusion_matrix(y_test, pr2))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
#from sklearn.metrics import confusion_matrix, classification_report
dtc=DecisionTreeClassifier(criterion='gini')
dtc.fit(train_data_array, y_train)
pr=dtc.predict(test_data_array)
decision_tree_score=dtc.score(test_data_array, y_test)
print(decision_tree_score)
print(classification_report(y_test, pr))
print(confusion_matrix(y_test, pr))
```

## Ensemble

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=20, random_state = 0)
rfc.fit(train_data_array, y_train)
pr5 = rfc.predict(test_data_array)
random_forest_score=rfc.score(test_data_array, y_test)
print(random_forest_score)
print(classification_report(y_test, pr5))
print(confusion_matrix(y_test, pr5))
```

### Gradient boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gdc = GradientBoostingClassifier(n_estimators = 100, random_state = 0)
gdc.fit(train_data_array, y_train)
pr2 = gdc.predict(test_data_array)
gradient_boosting_score = gdc.score(test_data_array, y_test)
print(gradient_boosting_score)
print(classification_report(y_test, pr5))
print(confusion_matrix(y_test, pr5))
```
